## Q2.計算總分及平均成績
讓使用者輸入三科成績後計算總分及平均。


```python
chin = int(input("請輸入國文成績:"))
math = int(input("請輸入數學成績:"))
eng = int(input("請輸入英文成績:"))
print("成績總分 : %3d，平均成績 : %.2f" % ((chin+math+eng),(chin+math+eng)/3))
```

## Q5.判斷成績等第
讓使用者輸入成績，若成績在 90 分以上就顯示「優等」， 80-89 分顯示「甲等」，70-79 分顯示「乙等」，60-69 分顯示「丙等」，60 分以下顯示「丁等」


```python
grade = int(input("請輸入成績"))

if (grade >= 90):
    print("優等")
elif (grade >= 80 and grade <90):
    print("甲等")
elif (grade >= 70 and grade <80):
    print("乙等")
elif (grade >= 60 and grade <70):
    print("丙等")
else:
    print("丁等")
```

## Q6.百貨公司折扣戰
讓顧客輸入購買金額，若金額在 100000 元以上就打八折，金額在 50000 元以上就打八五折，金額在 30000 元以上就打九折，金額在 10000 元以上就打九五折。


```python
money = int(input("請輸入購物金額:"))

if (money >= 100000):
    money *= 0.8
elif (money >= 50000):
    money *= 0.85
elif (money >= 30000):
    money *= 0.9
elif (money >= 10000):
    money *= 0.95

print(int(money),"元")
```
