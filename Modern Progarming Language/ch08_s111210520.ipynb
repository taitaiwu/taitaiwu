## Q1.下載 YouTube 影片
以 Pytube 模組下載指定的 YouTube 影片，並顯示各項影片資訊。



> 註: Pytube 模組已不再支援，故改使用 yt-dlp 模組

```python
import yt_dlp

ydl_opts = {
    "format" : "best",  
    "outtmpl" : "C:/Users/s1112/OneDrive/桌面/Desktop/Modern_Programin_Language/Homework/%(title)s.%(ext)s",
    "progress_hooks" : [lambda d: print("下載狀態:" + str(d['status'])) if d['status'] == 'finished' else None]
}

with yt_dlp.YoutubeDL(ydl_opts) as ydl:
    info = ydl.extract_info("https://youtu.be/eC1HXqkvA3A?si=UN3PY3G9GPpNiSCS", download=True)

    print("影片標題 :" + str(info['title']))
    print("影片格式 : " + str(info['format']))
    print("影片解析度 : " + str(info['resolution']))
    print("影片長度 : " + str(info['duration']) + "秒")
```

## Q2.改變標籤及按鈕元件文字內容
開始時顯示歡迎文字，按下按鈕後標籤文字變為按鈕計次文字，按鈕文字變為**回復原來文字！**；再按一次按鈕，標籤文字變為**按了 2 次文字**，按鈕文字回復為**按我**；繼續按下按鈕，標籤文字會不斷累積按鈕次數，按鈕文字則在 **按我** 及 **回復原來文字！** 反覆切換。  



```python
import tkinter as tk

counter = 0

def click1():
    global counter
    counter += 1

    textvar1.set("你按我" + str(counter) + "次了!")

    if (counter%2 == 0):
        textvar2.set("按我!")
    else:
        textvar2.set("回復原來文字!")

win = tk.Tk()

textvar1 = tk.StringVar()
label1 = tk.Label(win , textvariable=textvar1 , padx=40 , pady=15)
textvar1.set("歡迎光臨Tkinter!")
label1.pack()

textvar2 = tk.StringVar()
button1 = tk.Button(win , textvariable=textvar2 , command=click1 , padx=30 , pady=5)
textvar2.set("pass me")
button1.pack()

win.mainloop()
```

## Q3.密碼確認視窗
使用者可在文字編輯元件中輸入密碼，若輸入「1234」表示密碼正確，顯示歡迎登入訊息；若輸入的密碼錯誤，顯示修改密碼訊息。  



```python
import tkinter as tk

def click2():
    if (textvar1.get() == "1234"):
        textvar2.set("密碼正確，歡迎登入!")
    else:
        textvar2.set("密碼錯誤，請修正密碼!")
        

win = tk.Tk()

label1 = tk.Label(win , text="請輸入密碼:")
label1.pack()

textvar1 = tk.StringVar()
text1 = tk.Entry(win , textvariable=textvar1)
text1.pack()

button1 = tk.Button(win , text="登入" , command=click2)
button1.pack()

textvar2 = tk.StringVar()
label2 = tk.Label(win , textvariable=textvar2)
label2.pack()

win.mainloop()
```

## Q4.最喜愛的球類運動 (單選)
點選球類選項後，下方會顯示點選的項目，只能單選。
