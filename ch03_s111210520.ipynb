## Q5.判斷質數
讓使用者輸入一個大於 1 的整數，判斷該數是否為質數。



```python
a = int(input("請輸入大於１的整數："))
k = 1

for i in range (2 , (int(a ** 0.5)+1)):
    if (a%i == 0):
        k = 0
        break

if (k == 1):
    print(a , "是質數!")
else:
    print(a , "不是質數!")
```

## Q6.while 迴圈計算班級成績
小美是一位教師，請你以 while 迴圈方式為小美設計一個輸入成績的程式，如果輸入「-1」表示成績輸入結束，在輸入成績結束後顯示班上總成績及平均成績。



```python
counter = 1
sum = 0

while (True):
    x = int(input("請輸入第" + str(counter) + "位學生的成績 : "))

    if (x == -1):
        break
    else:
        sum += x
        counter += 1

print("本班級總成績 :" , sum , "分，平均成績 : %.2f 分" % (sum/(counter-1)))
```

## Q8.以串列計算班級成績
小明是一位教師，請為小明設計一個輸入成績的程式，學生成績需存入串列做為串列元素，如果輸入「-1」表示成績輸入結束，最後顯示班上總成績及平均成績。



```python
grade = []
counter = 1
sum = 0

while (True):
    x = int(input("請輸入第" + str(counter) + "位學生的成績 : "))

    if (x == -1):
        break
    else:
        grade.append(x)
        counter += 1

for i in grade:
    sum += i

print("本班級總成績 : " , sum , "分，平均成績 : %.2f 分" % (sum/(counter-1)))
```

## Q10.顯示字典內容
先建立 3 筆字典資料：「鍵」為學生姓名，「值」為學生成績，再以程式新增 2 筆資料，最後以 keys() 及 values() 方法顯示字典內容。



```python
d = {"林小明" : 85,
     "曾山水" : 93,
     "鄭美麗" : 67}

d["黃明品"] = 71
d["陳莉莉"] = 98

list1 = list(d.keys())
list2 = list(d.values())

for i in range (0 , len(list1)):
    print(list1[i] , "的成績為" , list2[i] , "分")
```

## Q11.顯示字典內容- items()
使用 items() 方法顯示字典內容更為方便，範例內容及執行結果與上一個範例相同。


```python
d = {"林小明" : 85,
     "曾山水" : 93,
     "鄭美麗" : 67}

d["黃明品"] = 71
d["陳莉莉"] = 98

list1 = d.items()

for i,j in list1:
    print(i , "的成績為" , j , "分")
```
