## Q1.攝氏溫度轉華氏溫度
輸入攝氏溫度，求華氏溫度。



```python
def convert (temp):
    temp *= 1.8
    temp += 32

    print("華氏溫度為: %.1f 度" % temp)

a = int(input("請輸入攝氏溫度:"))
convert(a)
```

### 延伸:華氏溫度轉換攝氏溫度

```python
def convert (temp):
    temp -= 32
    temp *= 5
    temp /= 9

    print("攝氏溫度為: %.1f 度" % temp)

a = int(input("請輸入華氏溫度:"))
convert(a)
```

## Q2.多數值加法器

```python
def calsum (list1):
    sum = 0
    list1 = [int(i) for i in list1]

    for i in list1:
        sum += i

    list1 = [str(i) for i in list1]
    output = " + ".join(list1)
    print("不定數目參數範例:")
    print(len(list1) , "個參數:" , output , "=" , sum)

list1 = input().split("+")
calsum(list1)
```

## Q4.內建函式
讓使用者輸入若干個正整數，以內建函式顯示最大數、最小數、總和及排序。



```python
list1 = []

while (True):
    a = int(input("請輸入正整數 (-1:結束):"))
    
    if (a == -1):
        break
    else:
        list1.append(a)

def calculate (num):
    print("共輸入" , len(num) , "個數")
    print("最大數為 :" , max(num))
    print("最小數為 :" , min(num))

    sum = 0
    for i in num:
        sum += i
    print("輸入數的總和為 :" , sum)

    num.sort()
    num.reverse()
    print("輸入數由大到小排序為 :" , num)

calculate(list1)
```

## Q6.大樂透中獎號碼
大樂透中獎號碼為 6 個 1 到 49 之間的數字加 1 個特別號：撰寫程式取得大樂透中獎號碼，並由小到大顯示方便對獎。



```python
import random

def lotto():
    list1 = random.sample(range(1,50) , 7)
    list2 = []

    for i in list1:
        list2.append(i)

    list2 = [str(i) for i in list2]
    output = ", ".join(list2)
    print("本期大樂透中獎號碼為 :" , output)
    print("本期大樂透特別號為 :" , list1[len(list1)-1])

lotto ()
```

## Q7.列印時間函式所有資訊
大賽看板上需顯示以中華民國年份表示的現在時刻，給比賽選手做為參考。請設計程式以時間模組列印以中華民國年份表示的現在時刻及節約時間資訊。



```python
import time

roc = time.localtime().tm_year - 1911
weekday = ['星期日' , '星期一' , '星期二' , '星期三 ' , '星期四' , '星期五' , '星期六']
indest = ['無' , '有']

print("中華民國" , roc , time.strftime("年 %m 月 %d 日 %H 點 %M 分 %S 秒") , weekday[int(time.strftime("%w"))])
print("今天是今年的第" , time.localtime().tm_yday , "天 , 此地" + str(indest[time.localtime().tm_isdst]) + "日光節約時間")
```
