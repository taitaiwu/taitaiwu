## Q2.大量圖片處理
將pic目錄中所有圖片，全部轉換為800*600大小，再另存為.bmp 檔(存在bmp_photo目錄)，同時也將全部圖片灰階處理後依編號存成gray001.jpg、gray002.jpg⋯等檔案(存在gray_photo 目錄)。


```python
from PIL import Image
from time import sleep
import glob
import PIL
import shutil
import os

def create(folder):
    if (os.path.isdir(folder)):
        shutil.rmtree(folder)
        sleep(1)
    
    os.mkdir(folder)

file = glob.glob("pic\\*.png") + glob.glob("pic\\*.jpg")
folder1 = "bmp_photo"
folder2 = "gray_photo"

create(folder1)
create(folder2)

for i,j in enumerate(file):
    image1 = Image.open(j)
    image2 = image1.resize((800 , 600), Image.Resampling.LANCZOS)
    path,filename = os.path.split(j)
    name,ftype = os.path.splitext(filename)
    image2.save(folder1 + "/" + name + ".bmp")

    image3 = image2.convert("L")
    image3.save(folder2 + "/" + "gray" + str(i).zfill(3) + ".jpg")

    image1.close()
```
