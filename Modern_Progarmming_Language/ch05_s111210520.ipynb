## Q1.輸入兩個正整數求和，捕捉輸入的錯誤
在下方的程式使用者可以輸入兩個正整數，求兩數之和。若輸入非數值資料，請使用try⋯except 捕捉發生的錯誤。



```python
try:
    a = int(input("請輸入第一個整數 : "))
    b = int(input("請輸入第二個整數 : "))

    print("r=" , a+b)
except:
    print("發生輸入非數值的錯誤!")
```

## Q2.捕捉非數值資料和除數為 0 的錯誤
在以下的程式中，當輸入兩個正整數，求兩數之餘數時，可以 try⋯except 捕捉多個發生的錯誤，包括輸入非數值資料和除數為 0 的錯誤，大家都同聲說讚。



```python
try:
    a = int(input("請輸入第一個整數 : "))
    b = int(input("請輸入第二個整數 : "))
    c = a%b
    
except ValueError:
    print("發生輸入非數值的錯誤!")

except ZeroDivisionError:
    print("發生 integer division or modulo by zero 的錯誤，包括分母為 0 的錯誤!")

else:
    print("兩數之餘數為" , c)

finally:
    print("一定會執行的程式區塊")
```
